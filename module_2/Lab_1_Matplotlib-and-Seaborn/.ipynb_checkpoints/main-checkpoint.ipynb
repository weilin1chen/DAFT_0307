#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here

features = ['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']
feature_labels = ['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']

for feature, label in zip(features, feature_labels):
    plt.figure()
    plt.scatter(data=df, x=feature,y='Steps')
    plt.xlabel(label)
    plt.ylabel('Steps')
    plt.title(f'Scatter Plot: Steps vs. {label}')
    plt.grid(True)
    plt.show()
```

#### Use the right visulalization to show the distribution of column `Age`.

```python
# your code here

h = sns.displot(data=df1, x="Age")
h.set(xlabel="Age", ylabel="Count")
h.fig.suptitle("Age Distribution")
plt.show()
```

#### Use subplots and plot the distribution of the `Age`  with bins equal to 10, 20 and 50.

```python
# your code here

Age = list(df1['Age'])

fig, (ax1,ax2,ax3) = plt.subplots(nrows=1, ncols=3, figsize=(12, 6))


ax1.hist(Age, bins=10)
ax1.set_title('Distribution of Age (Bins=10)')
ax1.set_xlabel('Age')
ax1.set_ylabel('Frequency')

ax2.hist(Age, bins=20)
ax2.set_title('Distribution of Age (Bins=20)')
ax2.set_xlabel('Age')
ax2.set_ylabel('Frequency')

ax3.hist(Age, bins=50)
ax3.set_title('Distribution of Age (Bins=50)')
ax3.set_xlabel('Age')
ax3.set_ylabel('Frequency')

plt.tight_layout()

plt.show()
```

#### How does the bin size affect your plot?

```python
"""
The value on the frequency axis decreases.
"""
```

#### Use seaborn to show the distribution of column `Age`.

```python
# your code here

fig, (ax1,ax2,ax3) = plt.subplots(nrows=1, ncols=3, figsize=(12, 6))

sns.histplot(data=df1, x=Age, bins=10, ax=ax1)
ax1.set_title("Age Distribution")
ax1.set(xlabel="Age", ylabel="Frequency")

sns.histplot(data=df1, x=Age, bins=20, ax=ax2)
ax2.set_title("Age Distribution")
ax2.set(xlabel="Age", ylabel="Frequency")

sns.histplot(data=df1, x=Age, bins=50, ax=ax3)
ax3.set_title("Age Distribution")
ax3.set(xlabel="Age", ylabel="Frequency")

plt.tight_layout()

plt.show()
```

#### We suspect that there is a linear relationship between `Fare` and `Age`. Use the right plot to show the relationship between these 2 features. There are 2 ways, please do it both ways.
**Hint**: Use matplotlib and seaborn.

```python
# Method 1 - matplotlib
# your code here

plt.scatter(x=df1['Fare'],y=df1['Age'])

fit = np.polyfit(df1['Fare'], df1['Age'], deg=1)
reg_line = np.poly1d(fit)

plt.plot(df1['Fare'], reg_line(df1['Fare']))

plt.xlabel('Fare')
plt.ylabel('Age')

plt.show()
```

```python
# Method 2 - seaborn
# your code here

sns.regplot(data=df1,x='Fare',y='Age')
plt.show()
```
